### Automated Email Scheduling

```python
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
import schedule
import time
from datetime import datetime

def send_scheduled_email():
    # Gmail SMTP settings
    smtp_server = 'smtp.gmail.com'
    smtp_port = 587
    gmail_user = 'mitwork100@gmail.com'
    gmail_password = 'twcg nxvk xnxy fvmc'

    # Email details
    to_email = 'srikaran@mit.edu'
    subject = 'Scheduled Email'
    html_body = '<p>{Gem_Resp}</p>'  # the body of the email is the response generated by Gemini and saved into Gem_Resp

    # Create MIMEMultipart object
    msg = MIMEMultipart()
    msg['From'] = gmail_user
    msg['To'] = to_email
    msg['Subject'] = subject

    # Attach HTML body to the email
    msg.attach(MIMEText(html_body, 'html'))

    # Establish a connection to the SMTP server
    with smtplib.SMTP(smtp_server, smtp_port) as server:
        server.starttls()
        server.login(gmail_user, gmail_password)

        # Send the email
        server.sendmail(gmail_user, to_email, msg.as_string())

    print(f"Email sent to {to_email} at {datetime.now()}")

# Schedule the email to be sent every day at a specific time (change as needed)
schedule.every().day.at("22:43").do(send_scheduled_email)

# Keep the script running to execute scheduled tasks
while True:
    schedule.run_pending()
    time.sleep(1)
```
