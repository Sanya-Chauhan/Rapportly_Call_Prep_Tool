```python
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.image import MIMEImage
import schedule
import time
import base64
from datetime import datetime

# Set your Gmail credentials
gmail_user = 'xxxxxxx@gmail.com'
gmail_password = 'xxxx xxxx xxxx xxxx'
to_email = 'srikaran@xxx.xxx'

def send_scheduled_email():
    # Gmail SMTP settings
    smtp_server = 'smtp.gmail.com'
    smtp_port = 587

    # Email details
    subject = 'Rapportly: Client Engagement Resources'

    # Read HTML content from a file
    with open('index.html', 'r') as file:
        html_body = file.read()

    # Create MIMEMultipart object
    msg = MIMEMultipart()
    msg['From'] = gmail_user
    msg['To'] = to_email
    msg['Subject'] = subject

    # Attach HTML body to the email
    
    
    with open('images/image-1.png', 'rb') as img_file:
        img_data = base64.b64encode(img_file.read()).decode('utf-8')   
        html_body = html_body.replace('<img-1>', img_data)
        
    # Encode image-2.png to base64 and embed in HTML
    with open('images/image-2.png', 'rb') as img_file:
        img_data = base64.b64encode(img_file.read()).decode('utf-8')
        html_body = html_body.replace('<img-2>', img_data)
        
    msg.attach(MIMEText(html_body, 'html'))
        
    # Establish a connection to the SMTP server
    with smtplib.SMTP(smtp_server, smtp_port) as server:
        server.starttls()
        server.login(gmail_user, gmail_password)

        # Send the email
        server.sendmail(gmail_user, to_email, msg.as_string())

    print(f"Email sent to {to_email} at {datetime.now()}")

# Schedule the email to be sent every day at a specific time (change as needed)
schedule.every().day.at("17:44").do(send_scheduled_email)

# Keep the script running to execute scheduled tasks
while True:
    schedule.run_pending()
    time.sleep(1)
```
